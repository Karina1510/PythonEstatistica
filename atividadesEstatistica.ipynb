```python
#Exercício 7: Análise de Vendas de Produtos

#Uma loja registrou o número de unidades vendidas de diferentes produtos ao longo de uma semana. Cada produto tem um código único e o número de unidades vendidas é registrado em uma lista. Escreva um programa em Python que:
#1. Calcule a média aritmética das unidades vendidas.
#2. Encontre a mediana das unidades vendidas.
#3. Identifique o produto mais vendido (moda).
import random
from statistics import median, mode, StatisticsError

print("lojas roupa chique da kaka")
print("Bem vinda ao sistema da loja, veja as vendas ")
print("-------cód-------produto-------")
print("-------01--------boné----------")
print("-------02------blusa de frio---")
print("-------03------calça jeans-----")

# Dados das vendas dos produtos
semana_1A10 = [("01", 35), ("02", 3), ("03", 21)]
semana_11A17 = [("01", 30), ("02", 7), ("03", 8)]
semana_18A25 = [("01", 20), ("02", 2), ("03", 6)]
semana_26A31 = [("01", 10), ("02", 10), ("03", 1)]

lista_semanas = semana_1A10 + semana_11A17 + semana_18A25 + semana_26A31 #capsulando todas as semanas numa lista única

#atravês de uma 'fôrmula' mostrar que quero contar os resultados das 'chaves'/código dos produtos 
bone = [item[1] for item in lista_semanas if item[0] == '01']
blusaFrio = [item[1] for item in lista_semanas if item[0] == '02']
calcaJeans = [item[1] for item in lista_semanas if item[0] == '03']

print(f"O código '01', (Boné) aparece vez(es): {bone}")
print(f"O código '02', (Blusa de Frio) aparece vez(es): {blusaFrio}")
print(f"O código '03', (Calça Jeans) aparece vez(es): {calcaJeans}")

def media(vendas):
    if not vendas:
        return "Sem dados para calcular a média."
    total = sum(vendas)
    quantidade = len(vendas)
    return total / quantidade

def mediana(vendas):
    if not vendas:
        return "Sem dados para calcular a mediana."
    return median(vendas)

def moda(vendas):
    if not vendas:
        return "Sem dados para calcular a moda."
    try:
        return mode(vendas)
    except StatisticsError:
        return "Não há moda ou há múltiplas modas."

while True:
    numeroEscolhaProdutos = int(input("Digite o número da escolha do produto - Boné (1), Blusa de Frio (2), Calça Jeans (3): "))
        
    if numeroEscolhaProdutos == 1:
        produto = bone
        nome_produto = "Boné"
    elif numeroEscolhaProdutos == 2:
        produto = blusaFrio
        nome_produto = "Blusa de Frio"
    elif numeroEscolhaProdutos == 3:
        produto = calcaJeans
        nome_produto = "Calça Jeans"
    else:
        print("Escolha inválida. Tente novamente.")
        continue
        
    print(f"Você escolheu {nome_produto} com valores: {produto}")

    numeroEscolha = int(input("Digite o número da Escolha da opção de média (1), mediana (2) e moda (3): "))
    if numeroEscolha == 1:
        print(f"A média das vendas do {nome_produto} é: {media(produto)}")
    elif numeroEscolha == 2:
        print(f"A mediana das vendas do {nome_produto} é: {mediana(produto)}")
    elif numeroEscolha == 3:
        print(f"A moda das vendas do {nome_produto} é: {moda(produto)}")
    else:
        print("Escolha inválida. Tente novamente.")
```

```python
#Exercício 19: Você tem uma lista de 1000 números inteiros aleatórios entre 1 e 1000. Pegue uma amostra de 50 números dessa lista. Calcule a média e o desvio padrão da amostra.
#Escreva o programa em Python.

import random, math

numerosAleatorios = list(random.randint(1, 100) for _ in range(50))
print (numerosAleatorios)
quantidade = len(numerosAleatorios)
soma = sum(numerosAleatorios)
media = (soma/quantidade)
#variança amostral para conseguir o desvio padrão
soma_dos_quadrados_das_diferencas = sum((num - media) ** 2 for num in numerosAleatorios)
variancia = (soma_dos_quadrados_das_diferencas / quantidade - 1)
desvio_padrao_amostral = math.sqrt (variancia)

print(f"essa é a média da lista misteriosa {media} e essa é o desvio padrão {desvio_padrao_amostral}")
```

```python
#Exercício 20: Considere uma população com 3 grupos (A, B, C) com tamanhos diferentes: A (100), B (200), C (300). Faça uma amostra estratificada onde você tira 10% de cada grupo. Escreva o programa em Python.

import random
populacaoA = (100)
populacaoB =  (200)
populacaoc =  (300)

# 0.1 é a porcentagem de 10% por cada população, já que 10 por 100 (100 /10) é 0.1

tamanho_amostra_A =  int(0.1 * populacaoA)
tamanho_amostra_B = int(0.1 * populacaoB)
tamanho_amostra_C = int(0.1 * populacaoc)

grupo_A = list(range(1, populacaoA + 1))
grupo_B = list(range(1, populacaoB + 1))
grupo_C = list(range(1, populacaoc + 1))


numeros_aleatorios_a = random.sample(grupo_A, tamanho_amostra_A)
numeros_aleatorios_b = random.sample(grupo_B, tamanho_amostra_B)
numeros_aleatorios_c = random.sample(grupo_C, tamanho_amostra_C)

print(f"essas são as amostras aleatorias {numeros_aleatorios_a} na quantidade de {tamanho_amostra_A}")
print(f"essas são as amostras aleatorias {numeros_aleatorios_b} na quantidade de {tamanho_amostra_B}")
print(f"essas são as amostras aleatorias {numeros_aleatorios_c} na quantidade de {tamanho_amostra_C}")

```
